```python
import nltk
import random
from nltk.corpus import names
```

```python
#feature extractor function
def gender_features(word):
    f = {};
    f["firstletter"] = word[0].lower()
    f["2ndletter"] = word[1].lower()
    f["lastletter"] = word[-1].lower()
    f["2ndlastletter"] = word[-2].lower()
    
    f["last2"] = word[-1].lower() + word[-2].lower()
    f["first2"] = word[0].lower() + word[1].lower()
    
    f["firstandlast"] = word[0].lower() + word[-1].lower()
    
    f["length"] = len(word)
    
    for letter in 'abcdefghijklmnopqrstuvwxyz':
        f["count(%s)" % letter] = word.lower().count(letter)
        f["has(%s)" % letter] = (letter in word.lower())
       
    return f
```

```python
import csv
with open('ssa_female.csv', 'rb') as f:
    reader = csv.reader(f)
    ssa_female = list(reader)
    
import csv
with open('ssa_male.csv', 'rb') as f:
    reader = csv.reader(f)
    ssa_male = list(reader)
```

```python
#prepare a list of examples 
names = ssa_female + ssa_male

random.shuffle(names)
```

```python
#create the training and test sets
featuresets= [(gender_features(n),g) for (n,g) in names]
train_set = featuresets[1500:]
test_set = featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print nltk.classify.accuracy(classifier,test_set)
```

```python
classifier.show_most_informative_features(30)
```

```python
errors = []

devtest_names = names[500:1500]

for (name,tag)in devtest_names:
    guess = classifier.classify(gender_features(name))
    if guess != tag:
        errors.append((tag,guess,name))
    
for (tag,guess,name) in sorted(errors): #doctest: +ELLIPSES +NORMALIZE_WHITESPACE
    print 'correct=%-8s guess=%-8s name=%-30s' %(tag,guess,name)
```
