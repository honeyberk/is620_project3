```python
import nltk
from nltk.corpus import names
import random
import pandas as pd
```

```python
def gender_features(word):
    return {'last_letter':word[-2]+word[-1]}
```

```python
names = ([(name,'male') for name in names.words('male.txt')]+ [(name,'female') for name in names.words('female.txt')])
```

```python
random.shuffle(names)
```

```python
names
```

```python
featuresets = [(gender_features(n),g) for (n,g) in names]
featuresets
```

```python
train_set,test_set = featuresets[500:], featuresets[:500]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('Neo'))
```

```python
classifier.classify(gender_features('Trinity'))
```

```python
print nltk.classify.accuracy(classifier,test_set)
```

```python
classifier.show_most_informative_features(5)
```

```python
names = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/unisex-names/unisex_names_table.csv")
```

```python
names
```

```python
male_list2 =   names.loc[names['male_share'] > names['female_share'],['name']]
male_list2['sex'] = 'male'

female_list2 =   names.loc[names['female_share'] > names['male_share'],['name']]
female_list2['sex'] = 'female'

trainList = pd.concat([male_list2,female_list2]).values.tolist()
```

```python
random.shuffle(trainList)
```

```python
featuresets2 = [(gender_features(n),g) for (n,g) in trainList]
featuresets2
```

```python
train_set2,test_set2 = featuresets2[459:], featuresets2[:459]
train_set_merge = train_set + train_set2
test_set_merge = test_set + test_set2
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set_merge)
```

```python
classifier.classify(gender_features('Neo'))
```

```python
classifier.classify(gender_features('Trinity'))
```

```python
print nltk.classify.accuracy(classifier,test_set_merge)
```

```python
classifier.show_most_informative_features(5)
```
